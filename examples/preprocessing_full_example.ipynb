## Examining the images

Both of these images have four bands: blue, green, red, and near-infrared.

However, they were acquired by different sensors aboard different satellites. As a result, they have different spatial resolution, spatial coverage, and spectral responses.

Any number of bands may be used, so long as the band order is the same between the images.

## Distribution of spectral values in the target and reference images

Showing the original distribution of spectral values for each band.

```python
# since the images are very large, a smaller sample of 100,000 pixels is used for each band
n = 100000

fig, axs = plt.subplots(1,4, figsize=(16,5))
fig.tight_layout()

for band in range(4):
    band_targ = target_arr[band,:,:]
    band_ref = ref_arr[band,:,:]
    sample_pix_targ = np.random.choice(band_targ[band_targ != 0].ravel(), n)  # dropping all zeros
    sample_pix_ref = np.random.choice(band_ref[band_ref != 0].ravel(), n)  # dropping all zeros
    # Exclude +/- 2% of the data for clarity
    axs[band].hist2d(sample_pix_targ, sample_pix_ref, cmap=plt.cm.jet, bins=(100,100),
                    range=[[np.percentile(sample_pix_targ, 2),np.percentile(sample_pix_targ, 98)],
                            [np.percentile(sample_pix_ref, 2),np.percentile(sample_pix_ref, 98)]])
    axs[band].set_title("band " + str(band+1))
    axs[band].set_xlabel("Target image")
    
axs[0].set_ylabel("Reference image")
```

## Executing the preprocessing pipeline

The processing pipeline runs through custom_utils.main.

<b>Detailed description of the arguments for custom_utils.main:</b>

- image_ref: (str) filepath of an image to use as a radiometric reference. Only GeoTiffs are supported.
- image_reg_ref: (str) filepath of an image to use as a geolocation reference. May be the same as image_ref, but not required. Only GeoTiffs are supported.
- image_targ: (str) filepath of an image that will be preprocessed. The original image will not be altered.
- allow_reg: (bool) True to allow image co-registration. Default False.
- view_radcal_fits: (bool) display plots showing the radiometric normalization results using the invariant pixels. Default True.
- src_nodataval: (float) no-data value in the input images. Will be deprecated. Default 0.0.
- dst_nodataval: (float) no-data value to write in the output images. Default 0.0.
- udm: (str) filepath to the Unusable Data Mask, if needed. Only PlanetScope UDMs are currently supported, it will work with any mask where 1.0 represents unusuable data.
- outdir: (str) directory to save all outputs. Will be created if it does not already exist.
- datatype_out: GDAL datatype to save all outputs. Default is gdal.GDT_UInt16.
- is_ps: (bool) if the target image is from the PlanetScope constellation, PS-specific functions will be applied.

The plots below show how the preprocessing method has made the spectral properties of the images much more similar. The original target image is shown in teal and the reference image is shown in purple. The final image is the unfilled black line. 

```python
# load the output from the preprocessing pipeline as a numpy array
with rasterio.open(final_img_path, 'r') as src:
    final_arr = src.read()
```

```python
# since the images are very large, a smaller sample of 100,000 pixels is used for each band
n = 100000

fig, axs = plt.subplots(1,4, figsize=(16,5))
#fig.tight_layout()

for band in range(4):
    band_targ = target_arr[band,:,:]
    band_ref = ref_arr[band,:,:]
    band_final = final_arr[band,:,:]
    sample_pix_targ = np.random.choice(band_targ[band_targ != 0].ravel(), n)  # dropping all zeros
    sample_pix_ref = np.random.choice(band_ref[band_ref != 0].ravel(), n)  # dropping all zeros
    sample_pix_final = np.random.choice(band_final[band_final != 0].ravel(), n)  # dropping all zeros
    # Exclude +/- 2% of the data for clarity
    axs[band].hist(sample_pix_targ, bins=20, density=True,
                   range=[np.percentile(sample_pix_targ, 2),np.percentile(sample_pix_targ, 98)],
                   color='teal', alpha=0.50, label='target')
    axs[band].hist(sample_pix_ref, bins=20, density=True,
                   range=[np.percentile(sample_pix_ref, 2),np.percentile(sample_pix_ref, 98)],
                   color='darkorchid', alpha=0.50, label='reference')    
    axs[band].hist(sample_pix_final, bins=20, density=True,
                   range=[np.percentile(sample_pix_final, 2),np.percentile(sample_pix_final, 98)],
                   color='black', histtype='step', label='final')
    axs[band].set_title("band " + str(band+1))
    axs[band].set_xlabel("Spectral values")

axs[0].legend(loc='upper right')
```
